# 下面是输出每个名字对应出现的次数

```python
import os

#fd = os.popen("python C:\\Users\\SiChen\\tensorflow_facnet\\facenet_master\\src\\classifier.py TRAIN E:\\tensorflow_facnet\\room_face_160 C:\\Users\\SiChen\\tensorflow_facnet\\facenet_master\\20170512-110547\\20170512-110547.pb C:\\Users\\SiChen\\tensorflow_facnet\\facenet_master\\20170512-110547\\my_classifier.pkl --batch_size 1000")
fd = os.popen("python C:\\Users\\SiChen\\tensorflow_facnet\\facenet_master\\src\\classifier.py CLASSIFY E:\\tensorflow_facnet\\room_test_face_160 C:\\Users\\SiChen\\tensorflow_facnet\\facenet_master\\20170512-110547\\20170512-110547.pb C:\\Users\\SiChen\\tensorflow_facnet\\facenet_master\\20170512-110547\\my_classifier.pkl --batch_size 1000")

logt = fd.read()

name_id = 0

print(logt)
print("\n --------------以下为本次人脸识别的统计结果-----------------\n")

for t in os.listdir("E:\\tensorflow_facnet\\room_test_face_160"):
    if "." not in t:
        name_id = name_id + 1
        alist = logt.split(t)
        num = len(alist)-1       #这个可以拿出去，进行统计该名字一上午出现的次数
        print(name_id,": ",t,"<------>",num)

fd.close()
```

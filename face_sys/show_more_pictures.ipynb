```python
#显示两张图片
import matplotlib.pyplot as plt  
from PIL import Image
  
img1 =Image.open('D:\\newfolder1\\picture_80_.png')  
img2 =Image.open('D:\\newfolder1\\picture_81_.png')  

plt.subplot(121)  
plt.imshow(img1)  
plt.axis("off")  

plt.subplot(122)  
plt.imshow(img2)  
plt.axis("off")

plt.show() 
```

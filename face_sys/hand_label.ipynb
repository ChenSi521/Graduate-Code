```python
kernel1 = np.array([-1.0,1.0])
kernel2 = np.array([1.0,-1.0])
kernel3 = np.array([[-1.0,0.0],[0.0,1.0]])
kernel4 = np.array([[0.0,-1.0],[1.0,0.0]])
kernel5 = np.array([[0.0,1.0],[-1.0,0.0]])
kernel6 = np.array([[1.0,0.0],[0.0,-1.0]])
kernel7 = np.array([[-1.0],[1.0]])
kernel8 = np.array([[1.0],[-1.0]])
kernel_2 = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
```

```python
#
#
from PIL import Image
import os
import shutil
from IPython.display import clear_output
import matplotlib.pyplot as plt


#work_path_read = "D:\\newfolder0\\"
#work_path_read = "D:\\newfolder1\\"
source_face_work_path = "D:\\newfolder22_120"     #
source_face_work_path1 = "D:\\newfolder11"

goal_face_work_path1 = "D:\\newfolder23\\clear_face"      #
goal_face_work_path2 = "D:\\newfolder23\\fuzzy_face"      #


if os.path.exists(source_face_work_path):
    print("")
else:
    print("")


pic_number = len(os.listdir(source_face_work_path))
print(type(pic_number))
print("pic_number = ",pic_number)
#face_img_paths = [os.path.join(face_work_path,img) for img in os.listdir(face_work_path)]
#print(face_img_paths)
tttt = 0
for i,face_name in enumerate(os.listdir(source_face_work_path)):  #
    print(i,face_name)
    tttt = tttt + 1
    #
    if os.path.exists("D:\\newfolder20\\clear_face\\"+ face_name):      #
        continue
    if os.path.exists("D:\\newfolder20\\fuzzy_face\\"+ face_name):
        continue
    image = Image.open(source_face_work_path +"\\"+ face_name)
    #image.show()
    plt.subplot(121)
    plt.imshow(image)
    plt.axis("off")
    #plt.show() 
    
    image_features = Image.open(source_face_work_path1 +"\\"+ face_name)
    plt.subplot(122)
    plt.imshow(image_features)
    plt.axis("off")
    plt.show()
    
    image1 = np.array(image) 
    ###################################
    image_gaussion = cv2.filter2D(image1,cv2.CV_16S,kernel_2)
    image_grad1 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel1)
    image_grad2 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel2)
    image_grad3 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel3)
    image_grad4 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel4)
    image_grad5 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel5)
    image_grad6 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel6)
    image_grad7 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel7)
    image_grad8 = cv2.filter2D(image_gaussion,cv2.CV_16S,kernel8)
        
    image2 = np.maximum(np.maximum(np.maximum(np.maximum(np.maximum(np.maximum(np.maximum(image_grad1,
                                                                                      image_grad2),image_grad3),image_grad4),image_grad5),image_grad6),image_grad7),image_grad8)- \
                                                                                    np.minimum(np.minimum(np.minimum(np.minimum(np.minimum(np.minimum(np.minimum(image_grad1,
                                                                                      image_grad2),image_grad3),image_grad4),image_grad5),image_grad6),image_grad7),image_grad8)
    print(image2.var())
    ################################3
        
    stalls = input("123-") 
    print("stalls",stalls)
    if not tttt%1:        #5,jupyter
        clear_output() 
        tttt = 0
    if stalls == str(1):           #inputstr()
        pure_name = face_name.split(".")[0]       #txt
        #print(pure_name)
        shutil.copy(source_face_work_path+"\\"+pure_name+".png",goal_face_work_path1)             #newfolder8
       
        print("")
    if stalls == str(2):
        pure_name = face_name.split(".")[0] 
        #print(pure_name)
        pure_name = face_name.split(".")[0]     
        shutil.copy(source_face_work_path+"\\"+pure_name+".png",goal_face_work_path2)              #newfolder9
       
        print("")
        
    if stalls == str(3):
        continue
 
```

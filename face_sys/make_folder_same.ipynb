```python
#这段代码，可以将txt文件夹和png文件夹的内容相同，删除两个文件夹中不一样的文件。

import os

#work_path1 = "D:\\newfolder5\\front_face"
#work_path2 = "D:\\newfolder7\\front_face"
work_path1 = "D:\\newfolder5_120\\side_face\\新建文件夹"
work_path2 = "D:\\newfolder7\\side_face"

if os.path.exists(work_path1):
    print("work_path1文件夹存在")
else:
    print("work_path1文件夹不存在")
if os.path.exists(work_path2):
    print("work_path2文件夹存在")
else:
    print("work_path2文件夹不存在")

for i,face_name in enumerate(os.listdir(work_path1)):
    name_split = face_name.split(".")
    #print(name_split)
    if os.path.exists(work_path2+"\\"+ name_split[0]+".txt"):      #这样就不怕中途停止了
        continue
    else:
        os.remove(work_path1+"\\"+ face_name)
        print("删除："+work_path1+"\\"+ face_name)
```

```python
#将1文件夹中和2文件夹不同的图片复制到3文件夹。
#通过拉普拉斯算子来判断其是否相同。
import os 
import shutil
import cv2
import numpy as np

work_path1 = "D:\\newfolder65\\clear_side_face"
work_path2 = "D:\\newfolder43\\side_face"
work_path3 = "D:\\newfolder66"

if os.path.exists(work_path1):
    print("work_path1文件夹存在")
else:
    print("work_path1文件夹不存在")
if os.path.exists(work_path2):
    print("work_path2文件夹存在")
else:
    print("work_path2文件夹不存在")
if os.path.exists(work_path3):
    print("work_path2文件夹存在")
else:
    print("work_path2文件夹不存在")
    
kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]]) 
for img in os.listdir(work_path1):
    image = cv2.imread(work_path1 + "\\"+img)
    dst1 = cv2.filter2D(image,cv2.CV_16S,kernel).var()
    
    image1 = cv2.imread(work_path2 + "\\"+img)
    dst2 = cv2.filter2D(image1,cv2.CV_16S,kernel).var()
    if dst1 == dst2:
        continue
    else:
        shutil.copy(work_path1+"\\"+img,work_path3)
        
```
